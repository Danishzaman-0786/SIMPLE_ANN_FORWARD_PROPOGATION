```python
#simple nural network
import torch as t
import torch.nn as nn
class Simple_nn(nn.Module):
  def __init__(self,features):
    super().__init__()
    #ask,what are you add more layer
    self.hidden_layer=int(input("Numbers of Hidden_layes"))
    if self.hidden_layer==1:
      self.neurons=int(input("Numbers of neurons"))
      if self.neurons==1:
        self.l1=nn.Linear(features,self.neurons)
        self.sig_1=nn.Sigmoid()
      else:
        self.l1=nn.Linear(features,self.neurons)
        self.Rellu=nn.ReLU()
        self.l2=nn.Linear(self.neurons,1)
        self.sig_1=nn.Sigmoid()
    elif self.hidden_layer==2:
        self.first_hidden_laye=int(input("first_hidden_laye_numbers of neurons"))
        #print(self.first_hidden_laye)
        self.neurons=self.first_hidden_laye
        #print(self.neurons)
      # if self.neurons==self.neurons:
        self.l1=nn.Linear(features,self.neurons)
        print(self.l1)
        self.Rellu=nn.ReLU()
        self.second_hidden_laye=int(input("second_hidden_laye_numbers of neurons"))
        self.neurons2=self.second_hidden_laye
        self.l2=nn.Linear(self.neurons,self.neurons2)
        self.Rellu1=nn.ReLU()
        self.l3=nn.Linear(self.neurons2,1)
        self.sig_1=nn.Sigmoid()
  def forward(self,x):
    if self.hidden_layer==1 and self.neurons==1:
      z1=self.l1(x)
      return self.sig_1(z1)
    elif self.hidden_layer==1 and self.neurons>1:
      z1=self.l1(x)
      relu1=self.Rellu(z1)
      l12=self.l2(relu1)
      y_pred=self.sig_1(l12)
      return y_pred
    elif  self.hidden_layer==2:
      z1=self.l1(x)
      relu1=self.Rellu(z1)
      z2=self.l2(relu1)
      relu2=self.Rellu1(z2)
      z3=self.l3(relu2)
      return self.sig_1(z3)
  def loss_(self,y,y_pred):
      eps=1e-7
      y_pred=t.clamp(y_pred,eps,1-eps)
      loss1=-(y*t.log(y_pred)+(1-y)*(1-t.log(y_pred)))
      loss1=loss1.mean()
      return loss1




     # else:
      ## self.Rellu=t.rellu()
      #self.neurons=self.second_hidden_laye











```

```python
x_train_t,x_test_t=x_train_t.to(t.float32),x_test_t.to(t.float32)
```

```python
lr=0.01
```

```python
y_train_t.shape
```

```python
model(x_train_t).shape
```

```python
model.loss_(y_train_t, model(x_train_t))
```

```python
epochs=2
```

```python
for i in range(epochs):
  #forward
  y_pred=model(x_train_t)
  #loss
  loss1=model.loss_(y_train_t,y_pred)
  #backpro..
  loss1.backward()
  with t.no_grad():
    model.l1.weight-=lr*model.l1.weight.grad
    model.l1.bias-=lr*model.l1.bias.grad
  # Zero gradients after update
  model.l1.weight.grad.zero_()
  model.l1.bias.grad.zero_()
  print(f'epoch:{i} loss:{loss1}')
```

```python
model(x_test_t)
```

```python
with t.no_grad():
  y_pred=model(x_test_t)
  y_pred=(y_pred>0.5).int()
print(y_pred)
```

```python
#accuracy
with t.no_grad():
  accuracy=(y_pred==y_train_t).float().mean()
print(f"accuracy:={accuracy}")
```
